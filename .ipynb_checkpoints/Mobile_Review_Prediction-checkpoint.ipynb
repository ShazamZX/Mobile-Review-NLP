```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import num2word
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.naive_bayes import MultinomialNB
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
%matplotlib inline
```

```python
df = pd.read_csv('Amazon Mobile Rating.csv')
```

```python
df.drop("Unnamed: 0", axis=1, inplace= True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
def get_name(str):
    index = str.find('(')
    if index != -1:
        return str[:(index)]
    return str
df['Product Name'] = df['Product Name'].apply(get_name)
```

```python

ax = sns.countplot(x= 'Product Name', data= df)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,15))
ax = sns.countplot(x= 'Product Name', data= df, hue='Rating')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.tight_layout()
plt.show()
```

```python
df = df.dropna()
```

```python
df.drop(columns=['Product Name','Review-Title'], inplace= True)
```

```python
from num2words import num2words
def clean_text(st):
    def convert_func(matchobj):
      m =  matchobj.group(0)
      return num2words(m)      
    res = re.sub('[0-9]+', convert_func, st)
    res = re.sub('[^a-zA-Z]',' ',res)
    return res.strip()
```

```python
df['Review content'] = df['Review content'].apply(clean_text)
```

```python
df['Rating'] = (df['Rating'] > 2).astype(int)
df.reset_index()
```

```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
def lemma(data):
    lt = WordNetLemmatizer()
    corpus = ''
    datax = data.split()
    impwords = ['no','nor','not','only','few','more','most','all']
    for i in datax:
        ch = i.lower()
        words = ch.split()
        words = [lt.lemmatize(word) for word in words if (word not in stopwords.words('english')) and (word not in impwords)]
        words = ''.join(words)
        if (words != ''):
            corpus = corpus + ' ' + words;
    return (corpus.strip())
```

```python
train=df.sample(frac=0.8,random_state=200)
test=df.drop(train.index)
```

```python
train.reset_index(drop=True)
test.reset_index(drop=True)
```

```python
corpus = train['Review content'].apply(func=lemma)
```

```python
res = corpus.to_numpy()
```

```python
tfidf = TfidfVectorizer(max_features= 5000 , ngram_range=(1,3))
X_train = tfidf.fit_transform(res).toarray()
```

```python
y_train = np.array(train['Rating'])
```

```python
model = MultinomialNB(alpha= 0.30000000000000004)
model.fit(X_train,y_train)
```

```python
corpus1 = test['Review content'].apply(lemma)
X_test = tfidf.transform(corpus1).toarray()
y_test = np.array(test['Rating'])
model.score(X_test, y_test)
```

```python
import pickle
pickle.dump(tfidf,open('vectorizer.pkl','wb'))
pickle.dump(model,open('MobileReviewModel.pkl','wb'))
```
